```python
import ot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
M_list = []
for k in range(1,11):
    to_load = "matrices_1/om_before_break_" + str(k) + ".csv"
    M_list.append(np.array(pd.read_csv(to_load)))

after_break = []
for k in range(1,11):
    to_load = "matrices_1/om_after_break_" + str(k) + ".csv"
    M_list.append(np.array(pd.read_csv(to_load)))
```

```python
n_samples = before_break[0].shape[0]
```

```python
p = ot.unif(n_samples)
q = ot.unif(n_samples)
```

```python
dist_list = []
for k in range(19):
    gw0, log0 = ot.gromov.gromov_wasserstein(
    M_list[k], M_list[k + 1], p, q, "square_loss", verbose=True, log=True)
    dist_list.append(np.round(log0["gw_dist"], 4))
```

```python
plt.plot(range(19), dist_list)
plt.show()
```
